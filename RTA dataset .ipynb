```python
#dropping columns which has more than 2500 missing valuesand time columns
df.drop(['Service_year_of_vehicle','Defect_of_vehicle','Work_of_casuality','Fitness_of_casuality','Time'],axis=1,inplace=True)
df.head()
```

```python
#sorting categorail columns to new variables
categorical=[i for i in df.columns if df[i].dtype=='O']
print('The categorical variables are',categorical)
```

```python
#for categorical values we can replace the null values with mode of it
for i in categorical:
    df[i].fillna(df[i].mode()[0],inplace=True)
```

**Handling Categorical values**

```python
df.dtypes
```

Since there are so many categorical values, we need to use feature selection. we need to perform label encoding before applying chi square analysis

```python
#import label encoding
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

#creating a new df from performing chi2 analysis
df1=pd.DataFrame()

#adding all the categorical columns except the output to new dataframe
for i in categorical:
    if i!='Accident_severity':
        df1[i]=le.fit_transform(df[i])
```

```python
df1.info()
```

```python
#correlation
plt.figure(figsize=(22,17))
sb.set(font_scale=1)
sb.heatmap(df1.corr(), annot=True)
```

```python
df1.head()
```

```python
#import chi2 test
from sklearn.feature_selection import chi2
f_p_values=chi2(df1,df['Accident_severity'])
```

```python
f_p_values
```

```python
f_p_values1=pd.DataFrame({'features':df1.columns,'Fscore': f_p_values[0],'Pvalues':f_p_values[1]})
f_p_values1
```

```python
#after evaluting we are removing lesser imp columns and starting to a new dataframe
df2=df.drop(['Owner_of_vehicle','Type_of_vehicle','Road_surface_conditions','Pedestrian_movement','Casualty_severity',
             'Educational_level','Day_of_week','Sex_of_driver','Road_allignment','Sex_of_casualty'],axis=1)
df2.head()
```

```python
#get dummies
dummy=pd.get_dummies(df2[['Age_band_of_driver', 'Vehicle_driver_relation', 'Driving_experience', 'Area_accident_occured', 'Lanes_or_Medians',
                         'Types_of_Junction', 'Road_surface_type', 'Light_conditions', 'Weather_conditions', 'Type_of_collision', 'Vehicle_movement',
                         'Casualty_class', 'Age_band_of_casualty', 'Cause_of_accident']],drop_first=True)
dummy.head()
```

```python
#concatinate dummy and old dataframe
df3=pd.concat([df2,dummy],axis=1)
df3
```

```python
#dropping dummies columns
df3.drop(['Age_band_of_driver', 'Vehicle_driver_relation', 'Driving_experience', 'Area_accident_occured', 'Lanes_or_Medians',
                         'Types_of_Junction', 'Road_surface_type', 'Light_conditions', 'Weather_conditions', 'Type_of_collision', 'Vehicle_movement',
                         'Casualty_class', 'Age_band_of_casualty', 'Cause_of_accident'],axis=1,inplace=True)
df3.head()
```

**Separting Independent and Dependent**

```python
x=df3.drop(['Accident_severity'],axis=1)
x.shape
```

```python
x.head()
```

```python
y=df3.iloc[:,2]
y.head()
```

```python
y.value_counts()
```

```python
#ploting countplot using seaborn
sb.countplot(x=y,palette='muted')
```

**Oversampling**

```python
#import SMOTE
from imblearn.over_sampling import SMOTE
oversample=SMOTE()
xo,yo=oversample.fit_resample(x,y)
```

```python
#checking the oversampling output
y1=pd.DataFrame(yo)
y1.value_counts()
```

```python
sb.countplot(x=yo,palette='muted')
```

**Splitting the Data**

```python
#splitting into testing and training data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(xo,yo,test_size=0.30,random_state=42)
```

```python
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

**KNN Model Creation**

**Prediction**

```python
from sklearn.neighbors import KNeighborsClassifier
model_KNN=KNeighborsClassifier(n_neighbors=5)
model_KNN.fit(x_train,y_train)
```

```python
y_pred=model_KNN.predict(x_test)
y_pred
```

**Checking Accurancy, Classification report and confusion matrix**

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score,ConfusionMatrixDisplay
```

Classification report

```python
report_KNN=classification_report(y_test,y_pred)
print(report_KNN)
```

```python
accuracy_KNN=accuracy_score(y_test,y_pred)
print(accuracy_KNN)
```

Confusion matrix

```python
matrix_KNN=confusion_matrix(y_test,y_pred)
print(matrix_KNN,'\n')
print(ConfusionMatrixDisplay.from_predictions(y_test,y_pred))
```
